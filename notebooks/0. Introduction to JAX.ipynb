**Introduction**

JAX is a Python library developed by Google for high performance numerical com-
putation and machine learning. It almost looks like Numpy and is sometimes referred
to as “NumPy on steroids”. Its core pillars are:

* Automatic Differentiation (AD): Computing gradients of arbitrary func-
tions
* Just-In-Time (JIT) Compilation: Compile Python code to GPU/TPU
using XLA
* Automatic Vectorization (vmap): Transform scalar functions to work on
batches
* Functional Programming: Pure functions enable composability
* Parallelization Across Devices (pmap): Distribute computations across
all GPUs at once

**Note:** JAX strictly relies on functional programming. Therefore the code must
be ”pure”, functions cannot have side effects, and arrays are immutable but instead,
you have to create a new array with the update.

**Why JAX for Machine Learning?**
* Performance: JIT compilation gives 10-100x speedups
* Research-Friendly: Easy to experiment with new algorithms
* Embedded ML: Lightweight, efficient inference on edge devices
* Composability: Stack transformations (grad–vma–pjit)
* Control: Fine-grained control over computation grap
